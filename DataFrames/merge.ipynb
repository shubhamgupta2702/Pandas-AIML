**Merging 2 Dataframes**

```python
import pandas as pd
import numpy as np
```

```python
employees={
  'employee_id':[1,2,3,4],
  'name':['Shubham','Vaishnavi','Vikram','Namrata'],
  'department':['SDE-2','SDE-3','IT','HR']
}

salaries={
  'employee_id':[1,2,5,6],
  'salary':[75000,55000,15000,35000],
  'Bonus':[15000,5600,7800,890]
}

employees = pd.DataFrame(employees)
salaries = pd.DataFrame(salaries)
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer')
```

```python
pd.merge(employees,salaries,how='left',on='employee_id')   #left -> employees
```

```python
pd.merge(employees,salaries,how='right',on='employee_id')   #right -> salaries
```
