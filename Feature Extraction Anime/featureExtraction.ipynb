**Q.1 - Make a new column for Episodes Count**

```python
def extract_episodes(txt):
  check = False
  data = ""
  for i in txt:
    if i == ')':
      check = False
      return data
    if check == True:
      data = data +i
    if i == '(':
      check = True
  
```

```python
df['Episodes'] = df["Title"].apply(extract_episodes)
```

**Q.2 - Make a new column for time stamps?**

```python
df.loc[0]['Title']
```

```python
def extract_timestamps(txt):
  check = False
  timestamps = ''
  for i in txt:
    if i == ',':
      check == False
      return timestamps
    if check == True:
      timestamps += i
    if i == ')':
      check = True
```

```python
df['Time Stamps'] = df['Title'].apply(extract_timestamps)
df['Time Stamps'] = df['Time Stamps'].str[:19]
```

```python
df
```

```python
from dateutil import parser  # For flexible date parsing
def calculate_months(date_str):
    start_str, end_str = date_str.split(' - ')
    start = parser.parse(start_str)
    end = parser.parse(end_str)
    return (end.year - start.year) * 12 + (end.month - start.month)
  
df['No. of Months'] = df['Time Stamps'].apply(calculate_months)
```
